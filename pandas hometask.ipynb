```python
import pandas as pd
```

```python
weather_hist = pd.read_csv('weatherHistory.csv')
```

```python
weather_hist.head()
```

```python
print('Самый жаркий день в Сегеде был', 
      weather_hist.loc[weather_hist['Temperature (C)'].idxmin()]['Formatted Date'], '.',
      'Температура составила', weather_hist['Temperature (C)'].min(), 'градусов.')
```

```python
print('Самый ветреный день в Сегеде был', 
      weather_hist.loc[weather_hist['Wind Speed (km/h)'].idxmax()]['Formatted Date'], '.',
      'Скорость ветра составила', weather_hist['Wind Speed (km/h)'].max(), 'км/ч.')
```

```python
weather_hist['Precip Type'].value_counts()
```

```python
weather_hist['Formatted Date'] = pd.to_datetime(weather_hist['Formatted Date']).apply(lambda x: x.date())
```

```python
weather_hist.head()
```

```python
weather_hist['year'] = pd.DatetimeIndex(weather_hist['Formatted Date']).year
```

```python
weather_hist['month'] = pd.DatetimeIndex(weather_hist['Formatted Date']).month
```

```python
rains_year = weather_hist.loc[weather_hist['year'] == 2014]
```

```python
rains_june = rains_year.loc[rains_year['month'] == 6]
```

```python
rains_july = rains_year.loc[rains_year['month'] == 7]
```

```python
rains_aug = rains_year.loc[rains_year['month'] == 8]
```
